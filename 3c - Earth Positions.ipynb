```python
import csv
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
# This grabs the position (x&y) array and tells what final distance is
# Also shows initial and final angular momenta
with open("PlanetsVV_2_0.005_escape.txt") as f:
    reader = csv.reader(f, delimiter="\t")
    d0_005 = list(reader)
pos0_005 = np.zeros((len(d0_005), 2))
vel0_005 = np.zeros((len(d0_005), 2))
i = 0
while i < len(d0_005):
    pos0_005[i] = float(d0_005[i][3]), float(d0_005[i][4])
    vel0_005[i] = float(d0_005[i][6]), float(d0_005[i][7])
    i += 1
finalDist = np.sqrt(pos0_005[len(d0_005)-1][0]**2 + pos0_005[len(d0_005)-1][1]**2)
print('The final distance from the sun is:', finalDist, 'AU')
momInit = (0.000003)*(np.sqrt(vel0_005[0][0]**2 + vel0_005[0][1]**2))*1
momFinal = (0.000003)*(np.sqrt(vel0_005[len(d0_005)-1][0]**2 + vel0_005[len(d0_005)-1][1]**2))*finalDist
print('Initial Angular Momentum: ', momInit)
print('Final Angular Momentum: ', momFinal)
```

```python
# This plots the y positions vs the x positions
fig = plt.figure(figsize=(5,5))
plt.plot(pos0_005[:,0], pos0_005[:,1])
plt.xlim([-1.5, 1.5])
plt.ylim([-1.5, 1.5])
plt.xlabel('X position (AU)')
plt.ylabel('Y position (AU)')
plt.grid()
plt.title('Earth Escaping Orbit (Time step = 0.005)')
```

```python
# This shows the initial and final energies to show if conserved or not
# Also takes their difference
with open("PlanetsVV_energy_2_0.005.txt") as f:
    reader = csv.reader(f, delimiter="\t")
    d0_005 = list(reader)
Energy0_005 = np.zeros((len(d0_005), 2))
i = 0
while i < len(d0_005):
    Energy0_005[i] = float(d0_005[i][2]), float(d0_005[i][3])
    i += 1
print("KE init:", Energy0_005[0][0], ", PE init:", Energy0_005[0][1])
print("KE final:", Energy0_005[len(d0_005)-2][0], ", PE final:", Energy0_005[len(d0_005)-2][1])
print("KE diff:", Energy0_005[len(d0_005)-2][0]-Energy0_005[0][0], ", PE diff: ", Energy0_005[len(d0_005)-2][1]-Energy0_005[0][1])
```

```python
# Here is a plot of Earth's orbit with a bigger time step
# You can see its orbit slowly increases, creating that thick-line donut shape.

with open("PlanetsVV_2_0.100.txt") as f:
    reader = csv.reader(f, delimiter="\t")
    d2 = list(reader)
pos2 = np.zeros((len(d2), 2))
i = 0
while i < len(d2):
    pos2[i] = float(d2[i][3]), float(d2[i][4])
    i += 1
fig = plt.figure(figsize=(5,5))
plt.plot(pos2[:,0], pos2[:,1])
plt.xlim([-1.5, 1.5])
plt.ylim([-1.5, 1.5])
plt.xlabel('X position (AU)')
plt.ylabel('Y position (AU)')
plt.title('Earth Orbit (Time step = 0.1)')
plt.grid()
```

```python
# Earth-Sun-Jupiter System, with three plots for three different masses of Jupiter.
# Also included is how far from 1 AU the Earth's final position deviates, and also
#  how far the Earth's final velocity deviates from its initial velocity of 2*Pi


# Jupiter Mass x 1
with open("PlanetsVV_3_0.005_J1.txt") as f:
    reader = csv.reader(f, delimiter="\t")
    J1 = list(reader)
    
earth_pos_J1 = J1[::2]
jupiter_pos_J1 = J1[1::2]
pos_E_J1 = np.zeros((len(earth_pos_J1), 2))
vel_E_J1 = np.zeros((len(J1), 2))
pos_J_J1 = np.zeros((len(jupiter_pos_J1), 2))
i = 0

while i < len(earth_pos_J1):
    pos_E_J1[i] = float(earth_pos_J1[i][3]), float(earth_pos_J1[i][4])
    vel_E_J1[i] = float(earth_pos_J1[i][6]), float(earth_pos_J1[i][7])
    pos_J_J1[i] = float(jupiter_pos_J1[i][3]), float(jupiter_pos_J1[i][4])
    i += 1

# Jupiter Mass x 10
with open("PlanetsVV_3_0.005_J10.txt") as f:
    reader = csv.reader(f, delimiter="\t")
    J10 = list(reader)
    
earth_pos_J10 = J10[::2]
jupiter_pos_J10 = J10[1::2]
pos_E_J10 = np.zeros((len(earth_pos_J10), 2))
vel_E_J10 = np.zeros((len(J10), 2))
pos_J_J10 = np.zeros((len(jupiter_pos_J10), 2))
i = 0

while i < len(earth_pos_J10):
    pos_E_J10[i] = float(earth_pos_J10[i][3]), float(earth_pos_J10[i][4])
    vel_E_J10[i] = float(earth_pos_J10[i][6]), float(earth_pos_J10[i][7])
    pos_J_J10[i] = float(jupiter_pos_J10[i][3]), float(jupiter_pos_J10[i][4])
    i += 1

# Jupiter Mass x 100
with open("PlanetsVV_3_0.005_J100.txt") as f:
    reader = csv.reader(f, delimiter="\t")
    J100 = list(reader)
    
earth_pos_J100 = J100[::2]
jupiter_pos_J100 = J100[1::2]
pos_E_J100 = np.zeros((len(earth_pos_J100), 2))
vel_E_J100 = np.zeros((len(J100), 2))
pos_J_J100 = np.zeros((len(jupiter_pos_J100), 2))
i = 0

while i < len(earth_pos_J100):
    pos_E_J100[i] = float(earth_pos_J100[i][3]), float(earth_pos_J100[i][4])
    vel_E_J100[i] = float(earth_pos_J100[i][6]), float(earth_pos_J100[i][7])
    pos_J_J100[i] = float(jupiter_pos_J100[i][3]), float(jupiter_pos_J100[i][4])
    i += 1    

# Setting up the figure size to be large enough for three plots
fig = plt.figure(figsize=(15,5))

# The first of three subplots
# Jupiter mass x 1
plt.subplot(131)
plt.plot(pos_E_J1[:,0], pos_E_J1[:,1])
plt.plot(pos_J_J1[:,0], pos_J_J1[:,1])
plt.xlim([-6, 6])
plt.ylim([-6, 6])
plt.xlabel('X position (AU)')
plt.ylabel('Y position (AU)')
plt.grid()
plt.title('Earth and Jupiter(1) Orbits')

# Jupiter mass x 10
plt.subplot(132)
plt.plot(pos_E_J10[:,0], pos_E_J10[:,1], label = 'Earth')
plt.plot(pos_J_J10[:,0], pos_J_J10[:,1], label = 'Jupiter')
plt.xlim([-6, 6])
plt.ylim([-6, 6])
plt.xlabel('X position (AU)')
plt.ylabel('Y position (AU)')
plt.legend(loc='upper left')
plt.grid()
plt.title('Earth and Jupiter(10) Orbits')

# Jupiter mass x 100
plt.subplot(133)
plt.plot(pos_E_J100[:,0], pos_E_J100[:,1])
plt.plot(pos_J_J100[:,0], pos_J_J100[:,1])
plt.xlim([-6, 6])
plt.ylim([-6, 6])
plt.xlabel('X position (AU)')
plt.ylabel('Y position (AU)')
plt.grid()
plt.title('Earth and Jupiter(100) Orbits')


# Printing the final position and velocity deviations for each mass of Jupiter
finalDist = np.sqrt(pos_E_J1[len(earth_pos_J1)-1][0]**2 + pos_E_J1[len(earth_pos_J1)-1][1]**2)
print('Final Earth Distance Deviation (Jupiter x 1):', np.abs(1-finalDist))
finalVel = np.sqrt(vel_E_J1[len(earth_pos_J1)-1][0]**2 + vel_E_J1[len(earth_pos_J1)-1][1]**2)
print('Final Earth Velocity Deviation (Jupiter x 1):', np.abs(2*np.pi-finalVel))

print('-------------------------------------------------------------------------')
finalDist = np.sqrt(pos_E_J10[len(earth_pos_J10)-1][0]**2 + pos_E_J10[len(earth_pos_J10)-1][1]**2)
print('Final Earth Distance Deviation (Jupiter x 10):', np.abs(1-finalDist))
finalVel = np.sqrt(vel_E_J10[len(earth_pos_J1)-1][0]**2 + vel_E_J10[len(earth_pos_J1)-1][1]**2)
print('Final Earth Velocity Deviation (Jupiter x 10):', np.abs(2*np.pi-finalVel))

print('-------------------------------------------------------------------------')
finalDist = np.sqrt(pos_E_J100[len(earth_pos_J100)-1][0]**2 + pos_E_J100[len(earth_pos_J100)-1][1]**2)
print('Final Earth Distance Deviation (Jupiter x 100):', np.abs(1-finalDist))
finalVel = np.sqrt(vel_E_J100[len(earth_pos_J1)-1][0]**2 + vel_E_J100[len(earth_pos_J1)-1][1]**2)
print('Final Earth Velocity Deviation (Jupiter x 100):', np.abs(2*np.pi-finalVel))
```
